# v4 Confidence Test on BSD10k

이 노트북은 추가 학습 없이 `outputs/confidence_filter_v4/predictions/BSD10k_oof_v4_scores.csv`를 사용해 BSD10k를 high-confidence binary task로 평가합니다.

중요한 구분:

- v4 best score인 `rank_average_binary_score`는 0/1 filtering score입니다. 이 값 자체는 1~5 class를 직접 예측하지 않습니다.
- 1~5 confusion matrix는 v4 입력에 포함된 v2 5-class 모델 출력(`prob_confidence_1~5`, `fiveclass_score`)을 기준으로 확인합니다.
- 1~5 예측을 binary로 접을 때는 `1,2,3 -> 0`, `4,5 -> 1` 규칙을 사용합니다.

평가 항목:

- v4 score threshold 기반 0/1 지표
- v2 5-class argmax/rounded score의 1~5 confusion matrix
- v2 5-class 예측을 0/1로 변환한 confusion matrix
- accuracy, MAE, MSE, precision, recall, F1

```python
from pathlib import Path
import json

import numpy as np
import pandas as pd

try:
    import matplotlib.pyplot as plt
    HAS_MATPLOTLIB = True
except ModuleNotFoundError:
    plt = None
    HAS_MATPLOTLIB = False

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    f1_score,
    mean_absolute_error,
    mean_squared_error,
    precision_score,
    recall_score,
)

try:
    display
except NameError:
    def display(obj):
        print(obj)

ROOT = Path.cwd()
if not (ROOT / "outputs").exists() and (ROOT.parent / "outputs").exists():
    ROOT = ROOT.parent

V4_SCORE_PATH = ROOT / "outputs" / "confidence_filter_v4" / "predictions" / "BSD10k_oof_v4_scores.csv"
V4_SUMMARY_PATH = ROOT / "outputs" / "confidence_filter_v4" / "reports" / "confidence_filter_v4_summary.json"
OUTPUT_DIR = ROOT / "outputs" / "v4_confidence_test"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

assert V4_SCORE_PATH.exists(), f"Missing file: {V4_SCORE_PATH}"
assert V4_SUMMARY_PATH.exists(), f"Missing file: {V4_SUMMARY_PATH}"

print("ROOT:", ROOT)
print("V4_SCORE_PATH:", V4_SCORE_PATH)
print("V4_SUMMARY_PATH:", V4_SUMMARY_PATH)
print("OUTPUT_DIR:", OUTPUT_DIR)
```

```python
# Load BSD10k v4 OOF scores
df = pd.read_csv(V4_SCORE_PATH)
summary = json.loads(V4_SUMMARY_PATH.read_text(encoding="utf-8"))

print("rows:", len(df))
print("columns:", df.columns.tolist())
print("v4 summary:")
print(json.dumps(summary, indent=2, ensure_ascii=False))
display(df.head())
```

```python
# Define targets and prediction scores
# Binary ground-truth rule: confidence 1,2,3 -> 0 / confidence 4,5 -> 1
true_confidence = df["confidence"].astype(int).to_numpy()
true_binary = (true_confidence >= 4).astype(int)

if "target_high_confidence" in df.columns:
    y_true = df["target_high_confidence"].astype(int).to_numpy()
    assert np.array_equal(y_true, true_binary), "target_high_confidence does not match confidence>=4 rule"
else:
    y_true = true_binary

# v4 best method from the report: Rank average: binary + expected score
best_method = summary.get("best_method", "Rank average: binary + expected score")
best_threshold = float(summary.get("best_threshold", 0.18))

method_to_col = {
    "Binary MLP probability": "binary_mlp_prob",
    "5-class P4+P5": "fiveclass_p45",
    "5-class expected score": "fiveclass_score",
    "5-class expected score scaled 0-1": "fiveclass_score_01",
    "5-class margin P45-P3": "fiveclass_margin_p45_minus_p3",
    "Rank average: binary + P45": "rank_average_binary_p45",
    "Rank average: binary + expected score": "rank_average_binary_score",
    "OOF logistic stacker": "logistic_stacker_prob",
}
v4_score_col = method_to_col.get(best_method, "rank_average_binary_score")
assert v4_score_col in df.columns, f"Missing v4 score column: {v4_score_col}"

v4_score = df[v4_score_col].astype(float).to_numpy()
y_pred_v4 = (v4_score >= best_threshold).astype(int)

# 5-class model component predictions.
# These are not v4's direct output; they are the v2 5-class predictions used as v4 inputs.
prob_cols = [f"prob_confidence_{i}" for i in range(1, 6)]
if all(col in df.columns for col in prob_cols):
    fiveclass_prob = df[prob_cols].to_numpy(dtype=float)
    fiveclass_pred_class_argmax = np.argmax(fiveclass_prob, axis=1) + 1
    y_pred_fiveclass_argmax_binary = (fiveclass_pred_class_argmax >= 4).astype(int)
else:
    fiveclass_prob = None
    fiveclass_pred_class_argmax = None
    y_pred_fiveclass_argmax_binary = None

if "fiveclass_score" in df.columns:
    fiveclass_score = df["fiveclass_score"].astype(float).to_numpy()
    fiveclass_pred_class_rounded = np.clip(np.rint(fiveclass_score), 1, 5).astype(int)
    y_pred_fiveclass_rounded_binary = (fiveclass_pred_class_rounded >= 4).astype(int)
else:
    fiveclass_score = None
    fiveclass_pred_class_rounded = None
    y_pred_fiveclass_rounded_binary = None

print("best_method:", best_method)
print("v4_score_col:", v4_score_col)
print("best_threshold:", best_threshold)
print("positive ratio true:", y_true.mean())
print("positive ratio v4 pred:", y_pred_v4.mean())
print("confidence distribution:")
display(pd.Series(true_confidence).value_counts().sort_index().rename_axis("confidence").reset_index(name="n"))
```

```python
def binary_metrics(y_true, y_pred, score=None):
    out = {
        "accuracy": accuracy_score(y_true, y_pred),
        "mae": mean_absolute_error(y_true, y_pred),
        "mse": mean_squared_error(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "pred_positive_rate": float(np.mean(y_pred)),
    }
    if score is not None:
        out["score_mae"] = mean_absolute_error(y_true, score)
        out["score_mse"] = mean_squared_error(y_true, score)
    return out

rows = []
rows.append({
    "method": f"v4: {best_method}",
    "score_col": v4_score_col,
    "threshold": best_threshold,
    **binary_metrics(y_true, y_pred_v4, v4_score),
})

if "binary_mlp_prob" in df.columns:
    binary_score = df["binary_mlp_prob"].astype(float).to_numpy()
    rows.append({
        "method": "v3 binary MLP @ 0.4",
        "score_col": "binary_mlp_prob",
        "threshold": 0.4,
        **binary_metrics(y_true, (binary_score >= 0.4).astype(int), binary_score),
    })

if fiveclass_score is not None:
    rows.append({
        "method": "v2 5-class expected score >= 4.0",
        "score_col": "fiveclass_score",
        "threshold": 4.0,
        **binary_metrics(y_true, (fiveclass_score >= 4.0).astype(int), np.clip((fiveclass_score - 1.0) / 4.0, 0, 1)),
    })
    rows.append({
        "method": "v2 rounded score: 1,2,3->0 / 4,5->1",
        "score_col": "round(fiveclass_score)",
        "threshold": np.nan,
        **binary_metrics(y_true, y_pred_fiveclass_rounded_binary),
    })

if "fiveclass_p45" in df.columns:
    p45 = df["fiveclass_p45"].astype(float).to_numpy()
    rows.append({
        "method": "v2 P4+P5 >= 0.5",
        "score_col": "fiveclass_p45",
        "threshold": 0.5,
        **binary_metrics(y_true, (p45 >= 0.5).astype(int), p45),
    })

if y_pred_fiveclass_argmax_binary is not None:
    rows.append({
        "method": "v2 argmax class: 1,2,3->0 / 4,5->1",
        "score_col": "argmax(prob_confidence_1~5)",
        "threshold": np.nan,
        **binary_metrics(y_true, y_pred_fiveclass_argmax_binary),
    })

metrics_df = pd.DataFrame(rows)
metrics_path = OUTPUT_DIR / "BSD10k_v4_binary_metrics.csv"
metrics_df.to_csv(metrics_path, index=False)
display(metrics_df)
print("saved:", metrics_path)
```

```python
def row_normalize(cm):
    return cm.astype(np.float32) / np.maximum(cm.sum(axis=1, keepdims=True), 1)


def show_binary_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
    cm_norm = row_normalize(cm)

    print(title)
    print("Confusion Matrix (counts)")
    print(pd.DataFrame(cm, index=["true_0_conf_1to3", "true_1_conf_4to5"], columns=["pred_0", "pred_1"]))
    print("\nNormalized Confusion Matrix (row-wise, 0~1)")
    display(pd.DataFrame(cm_norm, index=["true_0_conf_1to3", "true_1_conf_4to5"], columns=["pred_0", "pred_1"]))

    if HAS_MATPLOTLIB:
        plt.figure(figsize=(5, 4))
        plt.imshow(cm_norm, vmin=0, vmax=1)
        plt.title(title)
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.xticks([0, 1], ["0", "1"])
        plt.yticks([0, 1], ["0", "1"])
        for i in range(cm_norm.shape[0]):
            for j in range(cm_norm.shape[1]):
                plt.text(j, i, f"{cm_norm[i, j]:.2f}\n({cm[i, j]})", ha="center", va="center")
        plt.colorbar()
        plt.tight_layout()
        plt.show()
    else:
        print("matplotlib is not installed; skipping plot.")


def show_fiveclass_confusion_matrix(y_true_class, y_pred_class, title):
    labels = [1, 2, 3, 4, 5]
    cm = confusion_matrix(y_true_class, y_pred_class, labels=labels)
    cm_norm = row_normalize(cm)
    index = [f"true_{label}" for label in labels]
    columns = [f"pred_{label}" for label in labels]

    print(title)
    print("5-class Confusion Matrix (counts)")
    display(pd.DataFrame(cm, index=index, columns=columns))
    print("\n5-class Normalized Confusion Matrix (row-wise, 0~1)")
    display(pd.DataFrame(cm_norm, index=index, columns=columns))

    if HAS_MATPLOTLIB:
        plt.figure(figsize=(6, 5))
        plt.imshow(cm_norm, vmin=0, vmax=1)
        plt.title(title)
        plt.xlabel("Predicted confidence")
        plt.ylabel("True confidence")
        plt.xticks(range(5), labels)
        plt.yticks(range(5), labels)
        for i in range(cm_norm.shape[0]):
            for j in range(cm_norm.shape[1]):
                plt.text(j, i, f"{cm_norm[i, j]:.2f}\n({cm[i, j]})", ha="center", va="center")
        plt.colorbar()
        plt.tight_layout()
        plt.show()
    else:
        print("matplotlib is not installed; skipping plot.")


# v4 direct output is binary, so this is the direct v4 0/1 confusion matrix.
show_binary_confusion_matrix(
    y_true,
    y_pred_v4,
    f"v4 binary confusion matrix: {best_method} @ threshold {best_threshold:.3f}",
)
```

```python
# 5-class confusion matrices from the v2 5-class component used by v4.
# These answer: if the model predicts 1,2,3 -> binary 0 and 4,5 -> binary 1, how does it behave?

if fiveclass_pred_class_argmax is not None:
    show_fiveclass_confusion_matrix(
        true_confidence,
        fiveclass_pred_class_argmax,
        "v2 5-class argmax confusion matrix",
    )
    show_binary_confusion_matrix(
        y_true,
        y_pred_fiveclass_argmax_binary,
        "v2 5-class argmax converted to binary: 1,2,3->0 / 4,5->1",
    )
else:
    print("prob_confidence_1~5 columns are missing; skipping 5-class argmax comparison.")

if fiveclass_pred_class_rounded is not None:
    show_fiveclass_confusion_matrix(
        true_confidence,
        fiveclass_pred_class_rounded,
        "v2 expected score rounded confusion matrix",
    )
    show_binary_confusion_matrix(
        y_true,
        y_pred_fiveclass_rounded_binary,
        "v2 rounded expected score converted to binary: 1,2,3->0 / 4,5->1",
    )
else:
    print("fiveclass_score column is missing; skipping rounded-score comparison.")
```

```python
# Save row-level predictions
pred_out = df.copy()
pred_out["v4_score_col"] = v4_score_col
pred_out["v4_score"] = v4_score
pred_out["v4_threshold"] = best_threshold
pred_out["v4_pred_binary"] = y_pred_v4
pred_out["true_confidence"] = true_confidence
pred_out["true_binary_confidence"] = y_true

if fiveclass_pred_class_argmax is not None:
    pred_out["v2_predicted_confidence_class_argmax"] = fiveclass_pred_class_argmax
    pred_out["v2_argmax_binary"] = y_pred_fiveclass_argmax_binary

if fiveclass_pred_class_rounded is not None:
    pred_out["v2_predicted_confidence_class_rounded_score"] = fiveclass_pred_class_rounded
    pred_out["v2_rounded_score_binary"] = y_pred_fiveclass_rounded_binary

pred_path = OUTPUT_DIR / "BSD10k_v4_binary_test_predictions.csv"
pred_out.to_csv(pred_path, index=False)
print("saved:", pred_path)
display(pred_out.head())
```
