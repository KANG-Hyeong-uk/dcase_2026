# BSD10k vs 1.1BSD10k Metadata Difference Analysis

이 노트북의 목적은 두 metadata 파일 사이에서 **무엇이, 얼마나, 어떤 방향으로 바뀌었는지**를 확인하는 것입니다.

- Old metadata: `1.1BSD10k_metadata.csv`
- New metadata: `BSD10k_metadata.csv`

분석은 아래 질문 순서로 읽으면 됩니다.

1. 두 파일의 규모와 `sound_id` 구성은 같은가?
2. 같은 `sound_id`끼리 비교했을 때 어떤 feature가 많이 바뀌었는가?
3. `confidence` 값은 전체적으로 증가했는가, 감소했는가, 아니면 특정 점수대로 이동했는가?
4. `class`, `class_idx`, `class_top` 같은 taxonomy 정보가 바뀌었는가?
5. 제목, 태그, 설명 같은 텍스트 metadata가 얼마나 수정되었는가?

그래프는 예쁘게 보여주기 위한 것이 아니라, **어디를 먼저 자세히 봐야 하는지 우선순위를 정하기 위한 요약 도구**입니다. 자세한 샘플은 `metadata_diff_outputs` 폴더의 CSV로 저장됩니다.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_columns', 80)
pd.set_option('display.max_colwidth', 140)
pd.set_option('display.width', 160)

METADATA_DIR = Path(r'C:\Users\solok\Downloads\2.0metadata\metadata')
OLD_PATH = METADATA_DIR / '1.1BSD10k_metadata.csv'
NEW_PATH = METADATA_DIR / 'BSD10k_metadata.csv'

OUTPUT_DIR = Path('metadata_diff_outputs')
OUTPUT_DIR.mkdir(exist_ok=True)

FEATURES = [
    'sound_id', 'class', 'class_idx', 'class_top', 'confidence',
    'uploader', 'license', 'title', 'tags', 'description'
]

OLD_PATH, NEW_PATH, OUTPUT_DIR.resolve()
```

## 1. Load Metadata

### 이 섹션에서 확인하려는 것

두 CSV가 정상적으로 읽히는지, 그리고 우리가 비교하려는 컬럼들이 실제로 존재하는지 먼저 확인합니다. 이 단계에서 컬럼명이 다르거나 누락되어 있으면 뒤의 비교 결과는 의미가 없어집니다.

```python
old = pd.read_csv(OLD_PATH)
new = pd.read_csv(NEW_PATH)

print('1.1BSD10k shape:', old.shape)
print('BSD10k shape   :', new.shape)

display(old.head())
display(new.head())
```

```python
schema = pd.DataFrame({'feature': sorted(set(old.columns).union(new.columns))})
schema['in_1.1BSD10k'] = schema['feature'].isin(old.columns)
schema['in_BSD10k'] = schema['feature'].isin(new.columns)
schema['expected_feature'] = schema['feature'].isin(FEATURES)
display(schema)

missing_old = set(FEATURES) - set(old.columns)
missing_new = set(FEATURES) - set(new.columns)
assert not missing_old, f'1.1BSD10k missing columns: {missing_old}'
assert not missing_new, f'BSD10k missing columns: {missing_new}'
```

## 2. Basic Dataset Overview

### 이 섹션에서 확인하려는 것

두 metadata 버전이 같은 데이터셋인지 먼저 봅니다.

- `rows`: 전체 행 수가 달라졌는가?
- `unique_sound_id`: 실제 고유 음원 수가 달라졌는가?
- `duplicate_sound_id_rows`: 같은 `sound_id`가 중복으로 들어간 문제는 없는가?
- `only_in_*`: 한쪽에만 있는 음원이 있는가?

여기서 차이가 크면 이후 feature 차이는 단순 수정이 아니라 데이터 추가/삭제의 영향일 수 있습니다.

```python
def overview(df: pd.DataFrame, name: str) -> dict:
    return {
        'dataset': name,
        'rows': len(df),
        'columns': len(df.columns),
        'unique_sound_id': df['sound_id'].nunique(),
        'duplicate_sound_id_rows': int(df['sound_id'].duplicated().sum()),
        'unique_class': df['class'].nunique(),
        'unique_class_idx': df['class_idx'].nunique(),
        'unique_class_top': df['class_top'].nunique(),
        'missing_cells': int(df.isna().sum().sum()),
    }

overview_df = pd.DataFrame([
    overview(old, '1.1BSD10k'),
    overview(new, 'BSD10k'),
])
display(overview_df)
```

```python
id_old = set(old['sound_id'])
id_new = set(new['sound_id'])

id_summary = pd.DataFrame([
    {'category': 'common_sound_id', 'count': len(id_old & id_new)},
    {'category': 'only_in_1.1BSD10k', 'count': len(id_old - id_new)},
    {'category': 'only_in_BSD10k', 'count': len(id_new - id_old)},
])
display(id_summary)

only_old = old[old['sound_id'].isin(id_old - id_new)].copy()
only_new = new[new['sound_id'].isin(id_new - id_old)].copy()

only_old.to_csv(OUTPUT_DIR / 'only_in_1.1BSD10k.csv', index=False, encoding='utf-8-sig')
only_new.to_csv(OUTPUT_DIR / 'only_in_BSD10k.csv', index=False, encoding='utf-8-sig')

display(only_old.head())
display(only_new.head())
```

## 3. Align by `sound_id` and Compare Features

### 이 섹션에서 확인하려는 것

공통 `sound_id`만 남긴 뒤, 같은 음원에 대해 각 metadata feature가 바뀌었는지 비교합니다.

막대그래프는 **어떤 컬럼이 가장 많이 수정되었는지**를 빠르게 보기 위한 것입니다. 막대가 긴 feature는 새 BSD10k에서 정책적으로 많이 정리되었거나, 자동 처리/수작업 수정이 많이 들어간 후보입니다.

읽는 법:
- `changed_count`가 크면 해당 feature가 많은 음원에서 바뀐 것입니다.
- `changed_rate_pct`는 공통 음원 중 몇 퍼센트가 바뀌었는지입니다.
- 아래 `all_changed_feature_values.csv`에는 실제로 어떤 값이 어떤 값으로 바뀌었는지 저장됩니다.

```python
old_common = old[old['sound_id'].isin(id_old & id_new)].copy()
new_common = new[new['sound_id'].isin(id_old & id_new)].copy()

merged = old_common.merge(
    new_common,
    on='sound_id',
    how='inner',
    suffixes=('_1.1BSD10k', '_BSD10k'),
    validate='one_to_one',
)

print('matched rows:', len(merged))
display(merged.head())
```

```python
def values_equal(a: pd.Series, b: pd.Series) -> pd.Series:
    return a.fillna('<NA>').astype(str).eq(b.fillna('<NA>').astype(str))

compare_features = [c for c in FEATURES if c != 'sound_id']
feature_summary_rows = []

for feature in compare_features:
    old_col = f'{feature}_1.1BSD10k'
    new_col = f'{feature}_BSD10k'
    equal_mask = values_equal(merged[old_col], merged[new_col])
    changed = int((~equal_mask).sum())
    feature_summary_rows.append({
        'feature': feature,
        'same_count': int(equal_mask.sum()),
        'changed_count': changed,
        'changed_rate_pct': round(changed / len(merged) * 100, 4),
    })

feature_diff_summary = pd.DataFrame(feature_summary_rows).sort_values('changed_count', ascending=False)
display(feature_diff_summary)
feature_diff_summary.to_csv(OUTPUT_DIR / 'feature_diff_summary.csv', index=False, encoding='utf-8-sig')
```

```python
ax = feature_diff_summary.sort_values('changed_count').plot(
    x='feature', y='changed_count', kind='barh', figsize=(9, 5), legend=False
)
ax.set_title('Which metadata features changed most?')
ax.set_xlabel('Number of common sound_id rows changed')
ax.set_ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
changed_records = []
for feature in compare_features:
    old_col = f'{feature}_1.1BSD10k'
    new_col = f'{feature}_BSD10k'
    mask = ~values_equal(merged[old_col], merged[new_col])
    if mask.any():
        temp = merged.loc[mask, ['sound_id', old_col, new_col]].copy()
        temp.insert(1, 'feature', feature)
        temp = temp.rename(columns={old_col: 'value_1.1BSD10k', new_col: 'value_BSD10k'})
        changed_records.append(temp)

all_changed = pd.concat(changed_records, ignore_index=True) if changed_records else pd.DataFrame(
    columns=['sound_id', 'feature', 'value_1.1BSD10k', 'value_BSD10k']
)
all_changed.to_csv(OUTPUT_DIR / 'all_changed_feature_values.csv', index=False, encoding='utf-8-sig')
display(all_changed.head(30))
print('total changed feature cells:', len(all_changed))
```

## 4. Confidence Distribution Difference

### 이 섹션에서 확인하려는 것

`confidence` 점수의 전체 분포가 새 metadata에서 어떻게 바뀌었는지 확인합니다.

먼저 전체 confidence 값을 확인해서 예상 밖의 값이나 결측이 있는지 봅니다. 그 다음 `confidence` 1, 2, 3, 4, 5만 따로 뽑아 각 점수의 개수 차이를 비교합니다.

첫 번째 그래프는 두 버전의 점수별 개수를 나란히 보여줍니다. 두 번째 그래프는 `confidence` 1-5에 대해 `BSD10k - 1.1BSD10k` 개수 차이만 보여줍니다.

읽는 법:
- 차이 그래프에서 0보다 위면 새 BSD10k에서 그 confidence 점수가 늘어난 것입니다.
- 0보다 아래면 줄어든 것입니다.
- 예를 들어 낮은 confidence가 줄고 높은 confidence가 늘면, 새 metadata가 더 확신도 높은 샘플 중심으로 정리되었을 가능성을 볼 수 있습니다.

```python
old_conf = old['confidence'].value_counts(dropna=False).sort_index().rename('count_1.1BSD10k')
new_conf = new['confidence'].value_counts(dropna=False).sort_index().rename('count_BSD10k')

confidence_dist = pd.concat([old_conf, new_conf], axis=1).fillna(0).astype(int)
confidence_dist['diff_count_BSD10k_minus_1.1'] = confidence_dist['count_BSD10k'] - confidence_dist['count_1.1BSD10k']
confidence_dist['rate_1.1BSD10k_pct'] = (confidence_dist['count_1.1BSD10k'] / len(old) * 100).round(4)
confidence_dist['rate_BSD10k_pct'] = (confidence_dist['count_BSD10k'] / len(new) * 100).round(4)
confidence_dist['diff_rate_pp'] = (confidence_dist['rate_BSD10k_pct'] - confidence_dist['rate_1.1BSD10k_pct']).round(4)

confidence_scores = [1, 2, 3, 4, 5]
old_conf_numeric = pd.to_numeric(old['confidence'], errors='coerce')
new_conf_numeric = pd.to_numeric(new['confidence'], errors='coerce')

confidence_1_to_5_diff = pd.DataFrame(index=confidence_scores)
confidence_1_to_5_diff.index.name = 'confidence'
confidence_1_to_5_diff['count_1.1BSD10k'] = old_conf_numeric.value_counts().reindex(confidence_scores, fill_value=0).astype(int)
confidence_1_to_5_diff['count_BSD10k'] = new_conf_numeric.value_counts().reindex(confidence_scores, fill_value=0).astype(int)
confidence_1_to_5_diff['diff_count_BSD10k_minus_1.1'] = confidence_1_to_5_diff['count_BSD10k'] - confidence_1_to_5_diff['count_1.1BSD10k']
confidence_1_to_5_diff['rate_1.1BSD10k_pct'] = (confidence_1_to_5_diff['count_1.1BSD10k'] / len(old) * 100).round(4)
confidence_1_to_5_diff['rate_BSD10k_pct'] = (confidence_1_to_5_diff['count_BSD10k'] / len(new) * 100).round(4)
confidence_1_to_5_diff['diff_rate_pp'] = (confidence_1_to_5_diff['rate_BSD10k_pct'] - confidence_1_to_5_diff['rate_1.1BSD10k_pct']).round(4)

print('All confidence values, including unexpected or missing values:')
display(confidence_dist)
print('Confidence 1-5 count and rate differences:')
display(confidence_1_to_5_diff)

confidence_dist.to_csv(OUTPUT_DIR / 'confidence_distribution_diff.csv', encoding='utf-8-sig')
confidence_1_to_5_diff.to_csv(OUTPUT_DIR / 'confidence_1_to_5_count_diff.csv', encoding='utf-8-sig')
```

```python
plot_df = confidence_dist[['count_1.1BSD10k', 'count_BSD10k']]
ax = plot_df.plot(kind='bar', figsize=(8, 4))
ax.set_title('Did the overall confidence distribution change?')
ax.set_xlabel('confidence')
ax.set_ylabel('count')
ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()

ax = confidence_1_to_5_diff['diff_count_BSD10k_minus_1.1'].plot(kind='bar', figsize=(8, 3), color='tab:orange')
ax.axhline(0, color='black', linewidth=1)
ax.set_title('Confidence 1-5 count difference: BSD10k - 1.1BSD10k')
ax.set_xlabel('confidence')
ax.set_ylabel('BSD10k count minus 1.1BSD10k count')
ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Confidence Transition Table

### 이 섹션에서 확인하려는 것

분포만 보면 “몇 개가 늘고 줄었는지”는 알 수 있지만, **개별 음원의 confidence가 어디에서 어디로 이동했는지**는 알 수 없습니다. 그래서 같은 `sound_id` 기준으로 old confidence와 new confidence의 전이표를 만듭니다.

히트맵 읽는 법:
- 세로축은 1.1BSD10k의 confidence입니다.
- 가로축은 BSD10k의 confidence입니다.
- 대각선 값은 confidence가 그대로 유지된 음원 수입니다.
- 대각선 밖의 값은 confidence가 바뀐 음원 수입니다.
- 오른쪽 위 방향으로 값이 많으면 confidence가 올라간 경우가 많고, 왼쪽 아래 방향이면 내려간 경우가 많습니다.

```python
conf_old_col = 'confidence_1.1BSD10k'
conf_new_col = 'confidence_BSD10k'

confidence_transition = pd.crosstab(
    merged[conf_old_col],
    merged[conf_new_col],
    rownames=['1.1BSD10k confidence'],
    colnames=['BSD10k confidence'],
    dropna=False,
)
display(confidence_transition)
confidence_transition.to_csv(OUTPUT_DIR / 'confidence_transition_table.csv', encoding='utf-8-sig')

confidence_changed = merged.loc[
    ~values_equal(merged[conf_old_col], merged[conf_new_col]),
    ['sound_id', conf_old_col, conf_new_col, 'class_1.1BSD10k', 'class_BSD10k', 'title_1.1BSD10k', 'title_BSD10k']
].copy()
confidence_changed.to_csv(OUTPUT_DIR / 'confidence_changed_samples.csv', index=False, encoding='utf-8-sig')
display(confidence_changed.head(50))
print('confidence changed rows:', len(confidence_changed))
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(confidence_transition.values, cmap='Blues')
ax.set_xticks(range(len(confidence_transition.columns)))
ax.set_xticklabels(confidence_transition.columns)
ax.set_yticks(range(len(confidence_transition.index)))
ax.set_yticklabels(confidence_transition.index)
ax.set_xlabel('BSD10k confidence')
ax.set_ylabel('1.1BSD10k confidence')
ax.set_title('Where did individual confidence scores move?')

for i in range(confidence_transition.shape[0]):
    for j in range(confidence_transition.shape[1]):
        ax.text(j, i, confidence_transition.iloc[i, j], ha='center', va='center', fontsize=9)

fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
```

## 6. Class and Taxonomy Difference

### 이 섹션에서 확인하려는 것

모델 학습에서 가장 민감한 metadata는 label 계열입니다. 여기서는 `class`, `class_idx`, `class_top`이 바뀐 음원이 있는지 확인합니다.

이 값들이 바뀌었다면 단순 텍스트 정리가 아니라 학습 label 자체가 달라진 것이므로, 성능 비교나 재현 실험에서 특히 주의해야 합니다.

```python
taxonomy_cols = ['class', 'class_idx', 'class_top']

for col in taxonomy_cols:
    old_col = f'{col}_1.1BSD10k'
    new_col = f'{col}_BSD10k'
    changed = merged.loc[~values_equal(merged[old_col], merged[new_col]), ['sound_id', old_col, new_col]].copy()
    changed.to_csv(OUTPUT_DIR / f'{col}_changed_samples.csv', index=False, encoding='utf-8-sig')
    print(f'{col}: changed rows = {len(changed)}')
    display(changed.head(20))
```

```python
class_idx_mapping = (
    merged[['class_1.1BSD10k', 'class_idx_1.1BSD10k', 'class_BSD10k', 'class_idx_BSD10k']]
    .drop_duplicates()
    .sort_values(['class_1.1BSD10k', 'class_idx_1.1BSD10k', 'class_idx_BSD10k'])
)
class_idx_mapping['class_same'] = class_idx_mapping['class_1.1BSD10k'].astype(str).eq(class_idx_mapping['class_BSD10k'].astype(str))
class_idx_mapping['class_idx_same'] = class_idx_mapping['class_idx_1.1BSD10k'].astype(str).eq(class_idx_mapping['class_idx_BSD10k'].astype(str))

display(class_idx_mapping.head(80))
class_idx_mapping.to_csv(OUTPUT_DIR / 'class_idx_mapping_diff.csv', index=False, encoding='utf-8-sig')

print('unique class/index mapping rows:', len(class_idx_mapping))
print('class_idx mapping rows changed:', int((~class_idx_mapping['class_idx_same']).sum()))
```

## 7. Per-Class Confidence Difference

### 이 섹션에서 확인하려는 것

전체 confidence 분포가 바뀌었더라도, 그 변화가 모든 class에서 고르게 일어난 것인지 특정 class에 몰려 있는 것인지는 별개의 문제입니다.

이 표는 class별로 confidence 개수 차이를 보여줍니다. `abs_diff`가 큰 class는 confidence 수정이 집중된 class이므로 먼저 샘플을 확인할 후보입니다.

```python
old_class_conf = old.groupby(['class', 'confidence']).size().rename('count_1.1BSD10k')
new_class_conf = new.groupby(['class', 'confidence']).size().rename('count_BSD10k')

class_conf_diff = pd.concat([old_class_conf, new_class_conf], axis=1).fillna(0).astype(int).reset_index()
class_conf_diff['diff_count_BSD10k_minus_1.1'] = class_conf_diff['count_BSD10k'] - class_conf_diff['count_1.1BSD10k']
class_conf_diff = class_conf_diff.sort_values(['class', 'confidence'])

display(class_conf_diff.head(50))
class_conf_diff.to_csv(OUTPUT_DIR / 'class_confidence_distribution_diff.csv', index=False, encoding='utf-8-sig')

largest_class_conf_changes = class_conf_diff[class_conf_diff['diff_count_BSD10k_minus_1.1'] != 0].copy()
largest_class_conf_changes['abs_diff'] = largest_class_conf_changes['diff_count_BSD10k_minus_1.1'].abs()
display(largest_class_conf_changes.sort_values('abs_diff', ascending=False).head(30))
```

## 8. Text Metadata Difference

### 이 섹션에서 확인하려는 것

`uploader`, `license`, `title`, `tags`, `description`은 label은 아니지만 데이터 출처, 검색성, 설명 품질에 영향을 줍니다.

여기서는 텍스트 metadata가 얼마나 바뀌었는지 요약하고, 각 컬럼별 변경 샘플을 CSV로 저장합니다. 특히 `license` 변경은 사용 가능 조건과 연결될 수 있으므로 따로 확인할 가치가 있습니다.

```python
text_cols = ['uploader', 'license', 'title', 'tags', 'description']

text_change_summary = []
for col in text_cols:
    old_col = f'{col}_1.1BSD10k'
    new_col = f'{col}_BSD10k'
    mask = ~values_equal(merged[old_col], merged[new_col])
    text_change_summary.append({
        'feature': col,
        'changed_count': int(mask.sum()),
        'changed_rate_pct': round(mask.sum() / len(merged) * 100, 4),
    })
    sample = merged.loc[mask, ['sound_id', old_col, new_col]].copy()
    sample.to_csv(OUTPUT_DIR / f'{col}_changed_samples.csv', index=False, encoding='utf-8-sig')
    print(f'{col}: changed rows = {int(mask.sum())}')
    display(sample.head(10))

display(pd.DataFrame(text_change_summary).sort_values('changed_count', ascending=False))
```

## 9. One-Cell Summary

### 이 섹션에서 확인하려는 것

앞에서 본 핵심 숫자를 한 번에 복사하거나 보고서에 붙일 수 있게 요약합니다. 전체 결론을 쓸 때는 이 셀의 출력값을 먼저 보고, 변화가 큰 feature의 CSV 샘플을 확인하면 됩니다.

```python
summary_lines = []
summary_lines.append(f"1.1BSD10k rows: {len(old):,}, BSD10k rows: {len(new):,}")
summary_lines.append(f"Common sound_id: {len(id_old & id_new):,}")
summary_lines.append(f"Only in 1.1BSD10k: {len(id_old - id_new):,}")
summary_lines.append(f"Only in BSD10k: {len(id_new - id_old):,}")
summary_lines.append('')
summary_lines.append('Feature changes:')
for _, row in feature_diff_summary.iterrows():
    summary_lines.append(
        f"- {row['feature']}: {int(row['changed_count']):,} rows changed ({row['changed_rate_pct']:.4f}%)"
    )
summary_lines.append('')
summary_lines.append('Confidence distribution difference, BSD10k - 1.1BSD10k:')
for conf, row in confidence_dist.iterrows():
    summary_lines.append(
        f"- confidence {conf}: {int(row['diff_count_BSD10k_minus_1.1']):+,} rows ({row['diff_rate_pp']:+.4f} pp)"
    )
summary_lines.append('')
summary_lines.append(f"Saved detailed CSV files to: {OUTPUT_DIR.resolve()}")

print('\n'.join(summary_lines))
```

## 10. Interpretation of the Observed Differences

현재 결과를 보면 두 metadata 파일은 **같은 10,956개 음원에 대한 metadata 버전 차이**로 보는 것이 가장 자연스럽습니다.

핵심 해석:

- 행 수는 둘 다 10,956개입니다.
- 공통 `sound_id`도 10,956개입니다.
- 한쪽에만 있는 `sound_id`는 없습니다.
- 중복 `sound_id`도 없습니다.

따라서 BSD10k는 1.1BSD10k에 음원을 새로 추가하거나 삭제한 버전이라기보다, **동일한 음원 목록의 metadata를 일부 수정한 버전**입니다.

가장 크게 보이는 차이는 `class_idx`입니다. `class_idx`는 10,956개 전부 바뀌었지만, 이것은 모든 label이 바뀌었다는 뜻이 아닙니다. old 버전은 `0, 1, 2, ...`처럼 연속 번호를 썼고, new 버전은 `101, 102, 201, 301, 401, ...`처럼 상위 분류를 반영한 계층형 번호 체계를 씁니다. 예를 들어 `m-sp`는 `0 -> 101`, `m-si`는 `1 -> 102`, `is-p`는 `3 -> 201`, `fx-o`는 `11 -> 401`처럼 바뀝니다.

실제 class 문자열이 바뀐 음원은 58개입니다. 전체 10,956개 중 약 0.53%이므로, label 자체의 수정은 작지만 존재합니다. 그중 `class_top`까지 바뀐 경우는 33개입니다. 즉 58개 중 일부는 같은 상위 분류 안에서 세부 class만 바뀐 것이고, 33개는 상위 taxonomy까지 바뀐 더 큰 수정입니다.

`confidence`는 전체 분포 변화가 거의 없습니다.

| confidence | 1.1BSD10k | BSD10k | difference |
|---:|---:|---:|---:|
| 1 | 106 | 107 | +1 |
| 2 | 749 | 748 | -1 |
| 3 | 3280 | 3281 | +1 |
| 4 | 6045 | 6043 | -2 |
| 5 | 776 | 777 | +1 |

분포만 보면 거의 변화가 없어 보이지만, 같은 `sound_id` 기준으로 보면 confidence가 바뀐 음원은 39개입니다. 오른 음원이 20개, 내려간 음원이 19개라서 전체 방향성은 거의 중립입니다. 즉 BSD10k가 confidence를 전반적으로 올리거나 내린 버전이라기보다는, 일부 샘플의 confidence를 개별적으로 교정한 버전으로 보입니다.

또 중요한 점은 class가 바뀐 58개와 confidence가 바뀐 39개가 강하게 겹친다는 것입니다. 둘 다 바뀐 `sound_id`가 38개입니다. 즉 confidence 수정 대부분은 label/class 수정과 함께 일어난 것으로 보이며, 이는 단순한 점수 보정이라기보다 **애매하거나 잘못 분류된 샘플을 재검토하면서 class와 confidence를 같이 고친 경우**로 해석할 수 있습니다.

반대로 `uploader`, `license`, `title`, `tags`, `description`은 바뀐 행이 0개입니다. 따라서 출처, 라이선스, 제목, 태그, 설명 같은 텍스트 metadata는 그대로이고, 변화는 거의 전적으로 label/taxonomy/confidence 쪽에 집중되어 있습니다.

최종 결론:

BSD10k는 1.1BSD10k와 음원 구성은 동일하지만, `class_idx` 번호 체계를 계층형으로 전면 개편했고, 소수의 음원에 대해 `class`, `class_top`, `confidence`를 교정한 metadata 개정판입니다. 실험 관점에서는 파일 수나 음원 목록 차이보다 **label index mapping 변경**과 **58개 class label 수정**이 가장 중요한 차이입니다.

## 11. Class Change and Confidence Change Together

이 섹션은 `class`가 바뀐 58개 음원만 따로 보고, 각 음원이 **어떤 class에서 어떤 class로 바뀌었는지**와 그때 **confidence가 같이 어떻게 바뀌었는지**를 확인합니다.

읽는 법:

- `class_confidence_changed_samples`: 음원 단위 상세 목록입니다.
- `class_change_summary`: `class_old -> class_new` 전이별 개수와 confidence 변화 방향 요약입니다.
- `class_confidence_transition_summary`: `class_old -> class_new`와 `confidence_old -> confidence_new` 조합별 개수입니다.

`confidence_delta`가 양수면 confidence가 올라간 것이고, 음수면 내려간 것입니다.

```python
class_changed_mask = ~values_equal(merged['class_1.1BSD10k'], merged['class_BSD10k'])

class_confidence_changed_samples = merged.loc[
    class_changed_mask,
    [
        'sound_id',
        'class_1.1BSD10k', 'class_BSD10k',
        'class_top_1.1BSD10k', 'class_top_BSD10k',
        'class_idx_1.1BSD10k', 'class_idx_BSD10k',
        'confidence_1.1BSD10k', 'confidence_BSD10k',
        'title_1.1BSD10k',
    ]
].copy()

class_confidence_changed_samples = class_confidence_changed_samples.rename(columns={
    'class_1.1BSD10k': 'class_old',
    'class_BSD10k': 'class_new',
    'class_top_1.1BSD10k': 'class_top_old',
    'class_top_BSD10k': 'class_top_new',
    'class_idx_1.1BSD10k': 'class_idx_old',
    'class_idx_BSD10k': 'class_idx_new',
    'confidence_1.1BSD10k': 'confidence_old',
    'confidence_BSD10k': 'confidence_new',
    'title_1.1BSD10k': 'title',
})

class_confidence_changed_samples['confidence_delta'] = (
    pd.to_numeric(class_confidence_changed_samples['confidence_new'], errors='coerce')
    - pd.to_numeric(class_confidence_changed_samples['confidence_old'], errors='coerce')
)
class_confidence_changed_samples['confidence_change_direction'] = class_confidence_changed_samples['confidence_delta'].map(
    lambda x: 'up' if x > 0 else ('down' if x < 0 else 'same')
)
class_confidence_changed_samples['class_transition'] = (
    class_confidence_changed_samples['class_old'].astype(str)
    + ' -> '
    + class_confidence_changed_samples['class_new'].astype(str)
)
class_confidence_changed_samples['confidence_transition'] = (
    class_confidence_changed_samples['confidence_old'].astype(str)
    + ' -> '
    + class_confidence_changed_samples['confidence_new'].astype(str)
)

class_confidence_changed_samples = class_confidence_changed_samples[
    [
        'sound_id', 'class_transition', 'confidence_transition',
        'confidence_delta', 'confidence_change_direction',
        'class_old', 'class_new', 'class_top_old', 'class_top_new',
        'class_idx_old', 'class_idx_new', 'confidence_old', 'confidence_new', 'title',
    ]
].sort_values(['class_old', 'class_new', 'confidence_old', 'confidence_new', 'sound_id'])

display(class_confidence_changed_samples)
class_confidence_changed_samples.to_csv(
    OUTPUT_DIR / 'class_changed_with_confidence_samples.csv',
    index=False,
    encoding='utf-8-sig',
)

class_change_summary = (
    class_confidence_changed_samples
    .groupby(['class_old', 'class_new'], dropna=False)
    .agg(
        changed_sound_count=('sound_id', 'count'),
        confidence_changed_count=('confidence_delta', lambda s: int((s != 0).sum())),
        confidence_up_count=('confidence_delta', lambda s: int((s > 0).sum())),
        confidence_down_count=('confidence_delta', lambda s: int((s < 0).sum())),
        confidence_same_count=('confidence_delta', lambda s: int((s == 0).sum())),
        mean_confidence_delta=('confidence_delta', 'mean'),
    )
    .reset_index()
    .sort_values(['changed_sound_count', 'confidence_changed_count'], ascending=False)
)
class_change_summary['mean_confidence_delta'] = class_change_summary['mean_confidence_delta'].round(3)

display(class_change_summary)
class_change_summary.to_csv(
    OUTPUT_DIR / 'class_change_summary_with_confidence.csv',
    index=False,
    encoding='utf-8-sig',
)

class_confidence_transition_summary = (
    class_confidence_changed_samples
    .groupby(['class_old', 'class_new', 'confidence_old', 'confidence_new', 'confidence_delta'], dropna=False)
    .size()
    .reset_index(name='count')
    .sort_values(['count', 'class_old', 'class_new'], ascending=[False, True, True])
)

display(class_confidence_transition_summary)
class_confidence_transition_summary.to_csv(
    OUTPUT_DIR / 'class_confidence_transition_summary.csv',
    index=False,
    encoding='utf-8-sig',
)

print(f"class changed rows: {len(class_confidence_changed_samples)}")
print(f"class changed + confidence changed rows: {(class_confidence_changed_samples['confidence_delta'] != 0).sum()}")
print(f"confidence up among class-changed rows: {(class_confidence_changed_samples['confidence_delta'] > 0).sum()}")
print(f"confidence down among class-changed rows: {(class_confidence_changed_samples['confidence_delta'] < 0).sum()}")
print(f"confidence same among class-changed rows: {(class_confidence_changed_samples['confidence_delta'] == 0).sum()}")
```
